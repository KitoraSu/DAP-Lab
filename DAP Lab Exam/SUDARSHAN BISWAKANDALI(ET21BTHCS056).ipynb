# Loading the data (here we are using Pandas)
#### First, we need to load the data into a pandas DataFrame. This step involves reading the CSV file and checking the first few rows to understand the structure of the data.

```python
import pandas as pd

dff = pd.read_csv('Electric_Vehicle_Population_Data.csv')

dff
```

# DATA CLEANING

#### Before diving into analysis, it's crucial to clean the data. This includes handling missing values, correcting data types & possibly removing duplicates.

```python
print(dff.isnull().sum())

dff_cleaned = dff.dropna()

print("Data cleaning done.")
```

# DATA EXPLORATION
#### We'll explore several aspects of the data that could have market implications, such as the distribution of electric vehicles by make, model year & electric range.

## Number of EVs by Make

#### This visualization helps understand which car manufacturers are leading in electric vehicle production.
#### Here, the bar chart displays the number of electric vehicles by make, highlighting the market leaders in EV production.

```python
import matplotlib.pyplot as plt

vehicle_counts = dff_cleaned['Make'].value_counts()

plt.figure(figsize=(10, 6), facecolor='white')
vehicle_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Electric Vehicles by Make')
plt.xlabel('Make')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45)
plt.show()
```

## DISTRIBUTION OF EV BY MODEL YEAR

#### Understanding the trend of electric vehicle production over the years can indicate market growth.
#### This line graph shows the trend of electric vehicle production over the years, useful for identifying growth trends.

```python
year_counts = dff_cleaned['Model Year'].value_counts().sort_index()

plt.figure(figsize=(10, 6), facecolor='white')
year_counts.plot(kind='line', marker='o', color='green')
plt.title('Distribution of Electric Vehicles by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Number of Vehicles')
plt.grid(True)
plt.show()
```

## Market Impact Analysis

#### We'll analyze the data to find insights that could impact the market, such as identifying areas with high electric vehicle adoption.
#### This bar chart identifies cities with the highest number of electric vehicles, indicating urban areas where EVs are The most popular.

```python
top_cities = dff_cleaned['City'].value_counts().head(10)

plt.figure(figsize=(10, 6), facecolor='white')
top_cities.plot(kind='bar', color='purple')
plt.title('Top Cities for Electric Vehicle Adoption')
plt.xlabel('City')
plt.ylabel('Number of Electric Vehicles')
plt.xticks(rotation=45)
plt.show()
```
